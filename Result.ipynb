```python
#Get dataframe of results for experiment

def get_res(df):
  delta_list = df['delta'].unique().tolist()

  df_2 = pd.DataFrame()

  delta_output_list = []
  rejection_rate_list = []
  type_1_list = []
  power_list = []
  n_list = []

  for delta in delta_list:
    df_1 = df[df['delta'] == delta]
    val_count = df_1['result'].value_counts()
    rejection_rate = round((val_count[val_count.index == 'Reject Null Hypothesis for all k'][0])/(sum(val_count)),4)
    n = sum(df_1['sample_size'])/(df_1['sample_size'].count())
    
    for row, val in df_1.iterrows():
      rk_1 = val['rk_1']
      rk_2 = val['rk_2']
      res = val['result']
      req_count = 0
      req_count_2 = 0
      if res == 'Reject Null Hypothesis for all k':
        if rk_1 <= 1-delta or rk_2 <= 1-delta:
            req_count = req_count + 1
      elif res != 'Reject Null Hypothesis for all k':
        if rk_1 > 1-delta and rk_2 > 1-delta:
            req_count_2 = req_count_2 + 1

    power = 1-(req_count_2/len(df_1))
    type_1 = (req_count/len(df_1))

    delta_output_list.append(delta)
    type_1_list.append(type_1)
    rejection_rate_list.append(rejection_rate)
    power_list.append(power)
    n_list.append(n)

  df_2['n'] = np.array(n_list)
  df_2['delta'] = np.array(delta_output_list)
  df_2['type_1'] = np.array(type_1_list)
  df_2['rejection_rate'] = np.array(rejection_rate_list)
  df_2['power'] = np.array(power_list)

  return df_2   
```

```python
#Get dataframe of results for experiment 1

def get_res_1(df):
  prev_list = df['prevelance'].unique().tolist()
  delta_n_list = df['delta_n'].unique().tolist()
  delta_p_list = df['delta_p'].unique().tolist()

  df_2 = pd.DataFrame()

  prev_output_list = []
  delta_n_output_list = []
  delta_p_output_list = []
  rejection_rate_list = []
  type_1_list = []
  power_list = []
  n_list = []

  for prev in prev_list:
    for delta_n in delta_n_list:
      for delta_p in delta_p_list:
        df_1 = df[df['delta_n'] == delta_n][df['delta_p'] == delta_p][df['prevelance'] == prev]
        val_count = df_1['result'].value_counts()
        rejection_rate = round((val_count[val_count.index == 'Reject Null Hypothesis'][0])/(sum(val_count)),4)
        n = sum(df_1['sample_size'])/(df_1['sample_size'].count())

        #Type 1 error, Power calculation
        req_count = 0
        req_count_2 = 0
        for row, val in df_1.iterrows():
          sn_1 = val['sn_1']
          sn_2 = val['sn_2']
          sp_1 = val['sp_1']
          sp_2 = val['sp_2']
          dn = val['delta_n']
          dp = val['delta_p']
          res = val['result']
          delta_sn = round((sn_1-sn_2),4)
          delta_sp = round((sn_1-sn_2),4)

          if res == 'Reject Null Hypothesis':
            if delta_sn >= dn or delta_sp >= dp:
              req_count = req_count + 1
          if res != 'Reject Null Hypothesis':
            if delta_sn < dn and delta_sp < dp:
              req_count_2 = req_count_2 + 1

        power = 1-(req_count_2/len(df_1))
        type_1 = req_count/len(df_1)

        prev_output_list.append(prev)
        delta_n_output_list.append(delta_n)
        delta_p_output_list.append(delta_p)
        rejection_rate_list.append(rejection_rate)
        type_1_list.append(type_1)
        power_list.append(power)
        n_list.append(n)
  
  df_2['prev'] = np.array(prev_output_list)
  df_2['n'] = np.array(n_list)
  df_2['delta_n'] = np.array(delta_n_output_list)
  df_2['delta_p'] = np.array(delta_p_output_list)
  df_2['rejection_rate'] = np.array(rejection_rate_list)
  df_2['type_1'] = np.array(type_1_list)
  df_2['power'] = np.array(power_list)

  return df_2
```

```python
#Get dataframe of results for experiment 2

def get_res_2(df):
  prev_list = df['prevelance'].unique().tolist()
  delta_n_list = df['delta_n'].unique().tolist()
  delta_p_list = df['delta_p'].unique().tolist()

  df_2 = pd.DataFrame()

  prev_output_list = []
  delta_n_output_list = []
  delta_p_output_list = []
  delta_a_output_list = []
  rejection_rate_list = []
  type_1_list = []
  power_list = []
  n_list = []

  for prev in prev_list:
    for delta_n in delta_n_list:
      for delta_p in delta_p_list:
        df_1 = df[df['delta_n'] == delta_n][df['delta_p'] == delta_p][df['prevelance'] == prev]
        val_count = df_1['result'].value_counts()
        try:
          rejection_rate = round((val_count[val_count.index == 'Reject Null Hypothesis'][0])/(sum(val_count)),4)
        except:
          rejection_rate = 0
        n = sum(df_1['sample_size'])/(df_1['sample_size'].count())
        #Type 1 error, Power calculation
        sn1_list = []
        sn2_list = []
        sp1_list = []
        sp2_list = []
        
        req_count = 0
        req_count_2 = 0

        for row, val in df_1.iterrows():
          ac_1 = val['ac_1']
          ac_2 = val['ac_2']
          da = val['delta_a']
          res = val['result']
          delta_ac = round((ac_1-ac_2),4)
          if res == 'Reject Null Hypothesis':
            if delta_ac >= da:
              req_count = req_count + 1
          if res != 'Reject Null Hypothesis':
            if delta_ac < da:
              req_count_2 = req_count_2 + 1

        power = 1-(req_count_2/len(df_1))
        type_1 = req_count/len(df_1)

        prev_output_list.append(prev)
        delta_n_output_list.append(delta_n)
        delta_p_output_list.append(delta_p)
        delta_a_output_list.append(da)
        rejection_rate_list.append(rejection_rate)
        type_1_list.append(type_1)
        power_list.append(power)
        n_list.append(n)
  
  df_2['prev'] = np.array(prev_output_list)
  df_2['n'] = np.array(n_list)
  df_2['delta_n'] = np.array(delta_n_output_list)
  df_2['delta_p'] = np.array(delta_p_output_list)
  df_2['delta_a'] = np.array(delta_a_output_list)
  df_2['rejection_rate'] = np.array(rejection_rate_list)
  df_2['type_1'] = np.array(type_1_list)
  df_2['power'] = np.array(power_list)

  return df_2
```

```python
#Get dataframe of results for experiment 3

def get_res_3(df):
  delta_list = df['delta_d'].unique().tolist()

  df_2 = pd.DataFrame()

  prev_output_list = []
  delta_output_list = []
  rejection_rate_list = []
  type_1_list = []
  power_list = []
  n_list = []

  for delta in delta_list:
      df_1 = df[df['delta_d'] == delta]
      val_count = df_1['result'].value_counts()
      rejection_rate = round((val_count[val_count.index == 'Reject Null Hypothesis'][0])/(sum(val_count)),4)
      n = sum(df_1['sample_size'])/(df_1['sample_size'].count())

      #Type 1 error, Power calculation
      req_count = 0
      req_count_2 = 0
      
      for row, val in df_1.iterrows():
        ac_1 = val['P_10']
        ac_2 = val['P_01']
        da = val['delta_d']
        res = val['result']
        delta_ac = round((ac_1+ac_2),4)
        if res == 'Reject Null Hypothesis':
          if delta_ac >= da:
            req_count = req_count + 1
        if res != 'Reject Null Hypothesis':
          if delta_ac < da:
            req_count_2 = req_count_2 + 1

      power = 1-(req_count_2/len(df_1))
      type_1 = req_count/len(df_1)

      delta_output_list.append(delta)
      rejection_rate_list.append(rejection_rate)
      type_1_list.append(type_1)
      power_list.append(power)
      n_list.append(n)

  df_2['n'] = np.array(n_list)
  df_2['delta'] = np.array(delta_output_list)
  df_2['rejection_rate'] = np.array(rejection_rate_list)
  df_2['type_1'] = np.array(type_1_list)
  df_2['power'] = np.array(power_list)

  return df_2
```
